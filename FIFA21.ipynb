### Who are the ten highest rated player in the game?

```python
df_attributes.sort_values(by='OVA', ascending=False).head(10)
```

### How many players are considered `fast`, `medium` and `slow` paced in FIFA?

```python
def player_pace(pace):
    if pace > 82:
        return 'Fast'
    elif (pace < 82) & (pace > 75):
        return 'Medium'
    else:
        return 'Slow'
```

```python
labeled_pace = df_attributes['PAC'].apply(player_pace)
```

```python
print('Fast Rockets:',(labeled_pace  == 'Fast').sum())
print('Walking Folks:',(labeled_pace  == 'Medium').sum())
print('Crawling Turtles:',(labeled_pace  == 'Slow').sum())
```

### Finding outliers using IQR

```python
plt.figure(figsize=(8, 8))
df.boxplot(column=['OVA'])
plt.title('Box plot of numeric columns')
plt.xticks(rotation=45)
plt.show()
```

```python
Q1 = df['OVA'].quantile(0.25)
Q3 = df['OVA'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5*IQR
higher_bound = Q3 + 1.5*IQR
```

```python
higher_outliers = df.loc[(df['OVA'] > higher_bound), ['Name','OVA']]
higher_outliers
```

```python
lower_outliers = df.loc[(df['OVA'] < lower_bound), ['Name','OVA']]
lower_outliers
```
